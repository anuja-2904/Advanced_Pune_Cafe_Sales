```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['quantity_sold'], color='skyblue', edgecolor='black')
plt.title('Quantity sold Distribution')
plt.xlabel('Quantity Sold')
plt.ylabel('Frequency')
plt.show()
```

- All are sold equally

- The highest price is around 300
